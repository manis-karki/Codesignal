```python
def solution(symbol):
    if symbol in ['1','2','3','4','5','6','7','8','9','0']:
        return True
    else:
        return False
```


Determine if the given character is a digit or not.

Example

    For symbol = '0', the output should be
    solution(symbol) = true;
    For symbol = '-', the output should be
    solution(symbol) = false.
