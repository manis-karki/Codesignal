```python
import itertools
def overlap(read_a, read_b, min_length=2):
    start=0
    while True:
        start = read_a.find(read_b[:min_length],start)
        if start == -1:
            return 0
        if read_b.startswith(read_a[start:]):
            return len(read_a)-start
        start += 1
def shortestCommonSuperstring(string_set):
    shortest_sup = None
    for perm in itertools.permutations(string_set):
        sup = perm[0]
        for i in range(len(string_set)-1):
            olen= overlap(perm[i], perm[i+1], min_length=2)
            sup += perm[i+1][olen:]
        if shortest_sup is None or len(sup)< len(shortest_sup):
            shortest_sup = sup
    return shortest_sup
def max_overlap(reads):
    read_a, read_b = None, None
    olap_len=0
    for a,b in itertools.permutations(reads,2):
        olen= overlap(a,b, min_length=2)
        if olen > olap_len:
            read_a, read_b= a,b
            olap_len = olen
    return(read_a, read_b, olap_len)
def greedy_scs(reads):
    read_a, read_b, olen= max_overlap(reads)
    while olen>0:
        reads.remove(read_a)
        reads.remove(read_b)
        reads.append(read_a + read_b[olen:])
        read_a,read_b, olen = max_overlap(reads)
    return ''.join(reads)
##a=["hello","lloel", "elephant","phantom"]
final_seq= greedy_scs(Pool_Name)
```
